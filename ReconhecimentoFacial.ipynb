```python
#criar loop para ler todos os arquivos de male e todos embededs e concatenar numa lista
import os
folder_male = '/content/male'

lista_arquivos = []
male_features = []

for nome_arquivo in os.listdir(folder_male):
  caminho_arquivo = os.path.join(folder_male, nome_arquivo)
  if os.path.isfile(caminho_arquivo):
    deteccao_rosto = face_recognition.load_image_file(caminho_arquivo)
    coordenadas = face_recognition.face_locations(deteccao_rosto)
    male_features.append(face_recognition.face_encodings(deteccao_rosto, known_face_locations=coordenadas)[0])
    lista_arquivos.append(nome_arquivo)

print(male_features)
```

```python
#criar loop para ler todos os arquivos de male e todos embededs e concatenar numa lista
import os
folder_female = '/content/female'

lista_arquivos_female = []
female_features = []

for nome_arquivo in os.listdir(folder_female):
  caminho_arquivo = os.path.join(folder_female, nome_arquivo)
  if os.path.isfile(caminho_arquivo):
    deteccao_rosto = face_recognition.load_image_file(caminho_arquivo)
    coordenadas = face_recognition.face_locations(deteccao_rosto)
    female_features.append(face_recognition.face_encodings(deteccao_rosto, known_face_locations=coordenadas)[0])
    lista_arquivos_female.append(nome_arquivo)

print(female_features)
```

```python
#converter para dataframe e fazer o modelo
import pandas as pd

data_frame_males = pd.DataFrame(male_features)
data_frame_males['genero'] = 'M'
data_frame_males['arquivos'] = lista_arquivos
```

```python
data_frame_males
```

```python
#converter para dataframe e fazer o modelo

data_frame_females = pd.DataFrame(female_features)
data_frame_females['genero'] = 'F'
data_frame_females['arquivos'] = lista_arquivos_female
```

```python
data_frame_females
```

```python
#concatenando dois dataframes: male e female
data_set = pd.concat([data_frame_males, data_frame_females]).reset_index(drop = True)
```

```python
from sklearn.model_selection import train_test_split

from sklearn.svm import SVC

from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_set.drop(['genero', 'arquivos'], axis=1), data_set['genero'], test_size=0.3)
```

```python

model = SVC()

trained_model = model.fit(X_train, y_train)

results = trained_model.predict(data_set.drop(['genero', 'arquivos'], axis=1))
```

```python
results
```

```python
#descobrindo onde o modelo está errando e acertando
data_set['genero_pred'] = results
confusion_matrix(data_set['genero_pred'], data_set['genero'])
```
